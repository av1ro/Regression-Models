```python
headers={
    'Authorization': 'Bearer {}'.format(ACCESS_TOKEN), 
    'Accept': 'application/json'
    }

response = requests.get(url,
    params={'symbol': 'VIX', 'interval': 'daily', 'start': '2010-01-01', 'end': '2020-12-31'},
    headers=headers
)
json_response = response.json()
```

```python
df_vix_hist_price = pd.DataFrame(json_response['history']['day'])
```

```python
df_vix_hist_price.info()
```

```python
df_vix_hist_price['date']= pd.to_datetime(df_vix_hist_price['date'])
```

```python
df_vix_hist_price.info()
```

```python
df_vix_hist_price.tail(20)
```

```python
df_vix_hist_price['Week_Number'] = df_vix_hist_price['date'].dt.isocalendar().week
```

```python
df_vix_hist_price['year'] = df_vix_hist_price['date'].dt.isocalendar().year
```

```python
df_vix_hist_price["period"] = df_vix_hist_price["year"].astype(str) + df_vix_hist_price["Week_Number"].astype(str) 
```

```python
df_vix_hist_price['rank'] = df_vix_hist_price.groupby('period')['date'].rank(method='first')
```

```python
df_vix_hist_price.head()
```

```python
df_stock_hist_price['open_price_5'] = df_stock_hist_price.open.shift(4)
```

```python
df_stock_hist_price.head(10)
```

```python
df_stock_hist_price['percentage_change'] = abs(df_stock_hist_price['close'] - df_stock_hist_price['open_price_5'])*100/df_stock_hist_price['close']
```

```python
df_stock_hist_price.info()
```

```python
df_vix_hist_price.info()
```

```python
df_stock_hist_price_upd = df_stock_hist_price[4:]
```

```python
df_vix_hist_price_upd = df_vix_hist_price[4:]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
y = df_stock_hist_price_upd['percentage_change'].values
```

```python
X = df_vix_hist_price_upd['close'].values
```

```python
X
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,y)
plt.show()
```

```python
y_reshaped = y.reshape(-1,1)
X_reshaped = X.reshape(-1,1)
```

```python
reg = LinearRegression()
reg.fit(X_reshaped,y_reshaped)
```

```python
y_pred = reg.predict(X_reshaped)
```

```python
from matplotlib.pyplot import figure
figure(figsize=(12, 8), dpi=80)
plt.scatter(X,y)
plt.plot(X, y_pred,color='k')
plt.show()
```

```python
print(reg.score(X_reshaped, y_reshaped))
```
